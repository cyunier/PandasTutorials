# Visualization with Seaborn

"Seaborn provides an API on top of matplotlib which uses sane plot & color defaults, uses simple functions for common statistical plot types, and which integrates with the functionality provided by Pandas dataframes."
from: https://www.oreilly.com/learning/data-visualization-with-seaborn

```python
from __future__ import print_function, division

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), x, np.cos(x))
```

```python
# Let's look at the difference between Matplotlib & seaborn
import seaborn as sns
sns.set()
plt.plot(x, np.sin(x), x, np.cos(x));
```

## Histograms, KDE, and Densities

Exploring the seaborn API

```python
# Plotting histograms and joint distributions
data = np.random.multivariate_normal([0, 0], [[5,2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x','y'])

for col in 'xy':
    plt.hist(data[col], normed=True, alpha=0.5)
```

```python
# Smooth estimate of the distribution
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
# Histrogram and KDE combined:
sns.distplot(data['x'])
```
